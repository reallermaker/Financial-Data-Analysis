Indexing and selecting data are essential steps in data analysis. They help you extract the exact parts of your dataset that you need. Pandas, a popular Python library, offers many easy-to-use methods to quickly access and work with specific data

Indexing and selection are key in data analysis because they let you work with exactly the data you need.

**Access specific data** – You can quickly pick out a single value, certain rows or columns, or a combination of both.

**Filter and subset** – Apply conditions to keep only the rows or columns that match your criteria.

**Explore your data** – Focus on smaller parts of your dataset to spot patterns, trends, or unusual values.

**Manipulate data** – Update values, apply functions to certain sections, or create new features from existing data.

In short, indexing and selection help you retrieve, inspect, and modify data efficiently, making them essential for any analysis.





### Pandas Indexing and Selection

Pandas has a bunch of ways to get specific parts of your data, whether it's a single value, a row, or a whole section of a DataFrame. Let's look at the main methods.

#### **1. Square Brackets `[]`**

This is the most common way to select columns. You can use it to grab a single column by its name (which returns a Pandas Series) or multiple columns by passing a list of names (which returns a DataFrame).

-   To get one column: `df['column_name']`
-   To get multiple columns: `df[['col1', 'col2']]`

***

#### **2. Dot Notation `.`**

For columns that have simple names (no spaces or special characters), you can use dot notation. It's a quick and easy way to get a column, but it's less flexible than square brackets.

-   Example: `df.column_name`

***

#### **3. `.loc[]` (Label-Based Indexing)**

Use `.loc[]` when you want to select data based on the **names** of the rows and columns. It's great for slicing data with clear labels, and it's inclusive of the end value.

-   To select by row and column labels: `df.loc['row_label', 'column_label']`
-   To slice a range of rows and columns: `df.loc['row_start':'row_end', 'col_start':'col_end']`

***

#### **4. `.iloc[]` (Position-Based Indexing)**

Use `.iloc[]` when you want to select data based on the **integer positions** (like with a Python list or NumPy array), starting from 0. The slicing with `.iloc` is exclusive of the end position, just like in standard Python.

-   To select by row and column positions: `df.iloc[row_position, column_position]`
-   To slice a range of rows and columns: `df.iloc[0:5, 1:3]` (gets rows 0 through 4 and columns 1 through 2)

***

#### **5. Boolean Indexing**

This is a powerful way to filter data based on conditions. You create a series of `True`/`False` values (a "mask") and use it to select only the rows that meet your criteria.

-   Example: `df[df['age'] > 30]`

***

#### **6. Indexing with a Callable**

This is a more advanced technique where you pass a function to an indexer. The function takes the DataFrame or Series as input and returns a valid indexer (like a list of labels or a boolean mask). It's useful for creating reusable, complex selection logic, especially when you're chaining methods.

  at  , iat

When working with large datasets in Pandas, speed matters.  
If you need to read or update a single value in a DataFrame, the usual indexing methods can be slow.  
For faster access, Pandas provides two special methods:

- **`at`** – quick access for a single value by **label** (row/column names)  
- **`iat`** – quick access for a single value by **integer position** (row/column index)  

Use these when you only need one specific cell — they are much faster than regular indexing.


Pandas has two fast methods for getting or setting a single value in a DataFrame:

1. **`at`** – uses **labels** for row and column (label-based access)  
2. **`iat`** – uses **integer positions** for row and column (integer-based access)

These methods are much faster than regular indexing when working with just one cell.


````python
df = pd.DataFrame({'a': [1, 2, 3, 4], 'b': [4, 5, 6, 7]}, index=['A', 'B', 'C', 'D'])
df
````

'B' – row label

'a' – column label

It returns the value in row "B" and column "a"

````python
df.at['B', 'a']
````

Similarly, you can use **`at`** to set a single value:




````python
df.at['C', 'a'] = 999
df
````

The **`iat`** method works like **`at`**, but uses **integer positions** instead of labels.  
It lets you quickly get or set a single value by row and column index:




````python
df.iat[1, 0]
````

Similarly, you can use **`iat`** to set a single value by position:

````python
df.iat[1, 0] = 100
df
````

Using **`at`** and **`iat`** is faster than **`loc`**, **`iloc`**, or standard `[]` indexing when working with a single value.  
These methods are optimized for scalar access and avoid extra overhead from more flexible indexing.  

**Important:** The speed advantage is mainly for **single-cell operations**.  
If you need to access or modify **multiple values or slices**, **`loc`** or **`iloc`** are usually more suitable and still perform well.

<span style="color:#2ecc71; font-size:28px; font-weight:bold;">
   Boolean Indexing
</span>



Boolean indexing is a powerful way to **select rows or columns** in Pandas based on **conditions**.  
It lets you filter and extract subsets of data that meet specific criteria — a fundamental skill for data analysis.

Boolean indexing works by creating a **boolean mask**, which is an array of `True` and `False` values corresponding to each element in a Series or DataFrame.  

- **`True`** → the element is selected  
- **`False`** → the element is excluded  

When you apply this mask to a DataFrame or Series, only the `True` elements are returned.


In this example, the condition `series > 3` creates a **boolean mask**:

- `True` → elements greater than 4  
- `False` → elements less than or equal to 4 

Applying this mask to the Series selects only the values that satisfy the condition.


````python
series = pd.Series([1, 2, 3, 4, 5, 7, 8])
series > 4
````


- You can also create boolean masks for DataFrames by applying conditions to one or more columns.  
Only the rows where the condition is `True` will be selected.



````python
df = pd.DataFrame({'A': [1, 2, 3, 4, 5, 6, 7], 'B': [8, 9, 10, 11, 12, 13, 14]})
mask = df['A'] > 4
mask
````


<span style="color:#2ecc71; font-size:28px; font-weight:bold;">
   Selecting Rows and Columns Based on Boolean Conditions
</span>

You can use a boolean mask to filter rows in a DataFrame.


````python
df[mask]
````

- Combining Boolean Conditions Using Logical Operators
Boolean indexing lets you combine multiple conditions using logical operators: `&` (and), `|` (or), and `~` (not).

   

````python
mask = (df['A'] > 2) & (df['B'] <12)
df[mask]
````

````python
mask = (df['A'] > 8) | (df['B'] <12)
df[mask]
````


<span style="color:#cc462b; font-size:30px; font-weight:bold;">
   Indexing with isin
</span>



Pandas offers the `isin()` method to easily select rows from a Series or DataFrame based on a list of values. This method checks whether each element in a Series or DataFrame exists in a given set and returns a boolean mask that can be used for indexing.

`isin()` tests every element against the provided set and produces a boolean Series or DataFrame of the same shape. Here, `True` indicates that the element is present in the set, while `False` means it is not



````python
df = pd.DataFrame({
                   'A': [1, 2, 3, 4, 5, 6], 
                   'B': ['a', 'b', 'c', 'd', 'e', 'f']
})
df
````

- In this example, the `isin()` method is used on column `'A'` of the DataFrame `df`. It checks each value in `'A'` against the list `[3, 5]`. The resulting boolean mask is then used to filter the DataFrame, keeping only the rows where `'A'` is 3 or 5.


````python
values = [3, 5]
df[df['A'].isin(values)]
````

- The `isin()` method is useful for filtering data by a list of values. It helps you select or remove rows that match certain criteria.



- The `isin()` method is a simple but powerful way to select data that matches a list of specific values.

Imagine you have a big table of data (a DataFrame) and you only want to see the rows where a certain column contains a few specific items, like "apples" or "oranges". The `isin()` method checks if the value in each row is *in* your list of items.

It returns `True` for rows that match and `False` for those that don't, which is perfect for filtering.





- The `isin()` method helps you filter data in pandas easily. You can combine it with logical operators to create more specific conditions:

- **`&` (and):** Select rows that meet **all** conditions.  
  *Example:* Rows where a column value is in a list **and** another column is greater than a number.

- **`|` (or):** Select rows that meet **any** condition.  
  *Example:* Rows where a value is in one list **or** another list.

- **`~` (not):** Select rows that **do not** meet a condition.  
  *Example:* Rows where a column value is **not** in a list.

Using these, you can build custom filters to analyze your data efficiently.


````python
df = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': ['a', 'b', 'c', 'd', 'e'], 'C': [10, 20, 30, 40, 50]})
df
````

````python
df[(df['A'].isin([3, 5])) & (df['C'] > 33)]
````

- In this example, we use `isin()` to select rows where column `A` has the values 3 or 5.  
We combine this with another condition using `&` to select rows where column `C` is greater than 33.  

The final DataFrame includes only the rows that meet **both conditions**.

The `where()` method lets you **select or replace values** in a Series or DataFrame based on a boolean condition. It allows you to specify a condition (boolean mask) and decide which values to keep or replace. This makes it easy to perform conditional operations on your data. 

- **How it works:**  
  Values that satisfy the condition are **kept**, and values that do not are **replaced** (by default with `NaN`).


```python
series.where(cond, other=nan, inplace=False, axis=None, level=None, errors='raise')
dataframe.where(cond, other=nan, inplace=False, axis=None, level=None, errors='raise')


````python
s = pd.Series(range(1, 10)) 
s
````

````python
s.where(s > 2)
````

In this example, the `where()` method is used on the Series `s`.  
The condition `s > 2` makes a boolean mask: `True` for values greater than 2, and `False` otherwise.  
`selected_values` keeps the original numbers where the condition is `True` and replaces the rest with `NaN`.


````python
df = pd.DataFrame({'A': [1, 2, 30, 41, 5, 6], 'B': [100, 150, 300, 400, 550, 650]})
df.where(df >= 30)
````

The condition `df > 30` checks each element in `df`.  
It creates a new DataFrame, `selected_values`, where values greater than 20 are kept, and all others are replaced with `NaN`.


- Replacing Values Based on Conditions
The `where()` method can also replace values that don't meet a condition.  
By using the `other` parameter, you can set these values to a specific value or another Series/DataFrame.  

For example, you can replace values that do not satisfy the condition using `where()`:


````python
s = pd.Series(range(1, 10)) 
s.where(s > 5, 's')
````

In this example, `where()` is applied to the Series `s`.  
Values greater than 5 are kept, while all others are replaced with `s` using the `other` parameter.


- You can also set `other` to a Series or DataFrame.  
In this case, values that don't meet the condition are replaced with the matching elements from `other`.


````python
df = pd.DataFrame({'A': [1, 2, 30, 41, 5, 6], 'B': ['a', 'b', 'c', 400, 550, 650]})
mask = df['A'] > 15
df.where(mask, df['B'], axis=0)
````

The `where()` method returns a new DataFrame:  
- If the condition (mask) is True, the original value is kept.  
- If the condition is False, the value from `other` is used.  

The parameter `axis=0` applies the replacement column by column.   
- For rows where `A > 15` (30 and 41), the original values are kept.  
- For other rows, each column is replaced with the corresponding value from column `B`.

In **pandas**, the `merge()` function is used to **combine two DataFrames** based on one or more key columns.  


---

Parameters:

```python
merged_df = pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,
                     left_index=False, right_index=False, sort=False,
                     suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)
```

- `left`: first DataFrame  
- `right`: second DataFrame  
- `how`: type of merge  
  - `'inner'` → only rows with matching keys in both DataFrames  
  - `'left'` → all rows from the left DataFrame + matching rows from right  
  - `'right'` → all rows from the right DataFrame + matching rows from left  
  - `'outer'` → all rows from both DataFrames  
- `on`: name of the key column(s)



- how='inner' → keeps only rows that exist in both DataFrames.

````python
pd.merge(df, df1, on='customer_id', how='inner')
````

- how='outer' → keeps all rows from both DataFrames. If a customer_id exists in one DataFrame but not the other, the missing values are filled with NaN.

````python

pd.merge(df, df, on='customer_id', how='outer')
````

- cross
df → first DataFrame

Second df → second DataFrame (can be same as first)

how='cross' → performs a Cartesian product, combining every row of the first DataFrame with every row of the second DataFrame.

````python
pd.merge(df, df, how='cross')
````

**Pandas automatically detects NaN values, but we can also specify other values to be recognized as NAN**


Next, we will address removing rows where all values are NaN or where any of the row’s elements are NaN.


The `dropna()` function in pandas is used to remove rows or columns that contain missing values (NaN).

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

**axis:**

* `0` or `'index'` — Remove rows (default)
* `1` or `'columns'` — Remove columns

**how:**

* `'any'` — Remove the row or column if it contains **any** NaN value (default)
* `'all'` — Remove only rows or columns where **all** values are NaN

**thresh:**
The minimum number of non-NaN values required to keep the row or column.
For example, `thresh=2` means at least two non-NaN values must be present; otherwise, the row or column will be removed.

**subset:**
Check for NaN only in the specified columns.
Example: `subset=['col1', 'col2']`

**inplace:**

* If `True`, the changes are applied directly to the original DataFrame and nothing is returned.
* If `False` (default), a new DataFrame is returned.

دستور `dropna()` در pandas برای حذف سطرها یا ستون‌هایی استفاده می‌شود که شامل مقادیر گمشده (NaN) هستند.

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

**axis:**

* `0` یا `'index'` — حذف سطرها (پیش‌فرض)
* `1` یا `'columns'` — حذف ستون‌ها

**how:**

* `'any'` — اگر حتی یک مقدار NaN در سطر یا ستون وجود داشته باشد، حذف شود (پیش‌فرض)
* `'all'` — فقط سطر یا ستون‌هایی که همه مقادیرشان NaN است، حذف شوند

**thresh:**
حداقل تعداد مقادیر غیر NaN لازم برای نگه داشتن سطر یا ستون.
مثال: `thresh=2` یعنی حداقل دو مقدار غیر NaN وجود داشته باشد؛ در غیر این صورت، سطر یا ستون حذف می‌شود.

**subset:**
فقط در ستون‌های مشخص‌شده بررسی شود.
مثال: `subset=['col1', 'col2']`

**inplace:**

* اگر `True` باشد، تغییرات مستقیماً روی DataFrame اصلی اعمال می‌شود و خروجی برگردانده نمی‌شود.

*****CAGR(Compound Annual Growth Rate - CAGR)*****

$$CAGR = \left( \frac{V_{\text{final}}}{V_{\text{initial}}} \right)^{\frac{1}{t}} - 1$$

Suppose you invested 10 million dollers in a stock 4 years ago. Today, the total value of your shares has reached 25 million dollers.

**Question: What has been the average annual growth percentage of your investment?**

* **Initial Value ($V_{\text{initial}}$):** 10,000,000
* **Final Value ($V_{\text{final}}$):** 25,000,000
* **Number of years ($t$):** 4

The Compound Annual Growth Rate (CAGR) of this investment is **25.74%.**

**This means that your investment's performance over these 4 years was the equivalent of having earned a constant and continuous 25.74% compound return each year.**






Now, we are going to calculate the CAGR for the Mobarakeh Steel Company.

````python
price_2025 = yearly_prices.loc['2025-03-20', 'Year_End_Close_Price']
price_2023 = yearly_prices.loc['2023-03-20', 'Year_End_Close_Price'] 
t2 = int(2)
cagr2 = (price_2025/price_2023)**(1/t2) - 1 
print(f"The two-year CAGR is:{cagr2:.2%}")
````
